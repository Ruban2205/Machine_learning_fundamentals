```python
# Import the libraries 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Dataset/Iris.csv")
df.head(10)
```

### Find the Categorical, Numerical using Numpy

```python
# Use info to find Data type information 
df.info()
```

```python
# Using numpy to get the numerical data.
numerical_data = df.select_dtypes(include=np.number)
numerical_data.head(5)
```

```python
# Use numpy to get the categorical data
categorical_data = df.select_dtypes(include=np.object_)
categorical_data.head(5)
```

### Find the five number summary
- Quantile(0.25)
- Quantile(0.50)
- Quantile(0.75)
- Minimum and Maximum

```python
## Drop null values before calculating the datas
df = numerical_data.dropna()
df
```

```python
# Quantile
Q1 = np.quantile(df, 0.25)
Q2 = np.quantile(df, 0.50)
Q3 = np.quantile(df, 0.75)
print(Q1)
print(Q2)
print(Q3)
```
